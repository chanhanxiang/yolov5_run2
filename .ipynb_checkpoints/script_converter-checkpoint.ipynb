<h1>Script converter</h1>

Pre-labelled datasets can be downloaded from Roboflow. While there is an option for pre-labelled images to be exported in YOLO format for Roboflow, it was found that the exported annotations did not conform to YOLOv5 annotation format. An example is as follows:

<h5>From Roboflow</h5>

202 234 209 234 209 269 202 269 person 0

211 231 218 231 218 263 211 263 person 0

On the other hand, conventional YOLO format exported from labelimg looks like this:

<h5>From labelimg</h5>

0 0.532150776053215 0.4035476718403547 0.05543237250554327 0.059866962305986676

1 0.4994456762749445 0.725609756097561 0.07871396895787136 0.3093126385809313

YOLOv5 format indicates the class in number, followed by x_min, y_min, x_max and y_max. YOLO annotations exported from Roboflow has eight numbers, two of which appear repeating in random order. Given the absence of standardised guides on the Roboflow variant of YOLO annotations, it is difficult to use Roboflow's annotation for appropriate pre-processing.

There is also the option to export in XML format to get the x_min, y_min, x_max and y_max annotations from Roboflow, but this would require additional pre-processing. If the dataset is downloaded from Roboflow and in XML format, users can consider using the conversion code blocks outlined below if necessary. Otherwise, users can go over directly to the training section.

The following notebook presents the following segments:

- Import libraries

- XML to YOLO conversion

<h2>Preprocessing</h2>

<h3>Extract from XML</h3>

The following steps were performed in sequence:

- Assign variable to location where images and annotations are found

- Create function to extract x_min, y_min, x_max, y_max and class information from each annotation file

- Perform sample reading of contents of one annotation file


```python
main_dir = '/home/han/Documents/CV models/CV_Yolo5_2/main/'
```

```python
# Function to get the data from XML Annotation
def extract_info_from_xml(xml_file):
    root = ET.parse(xml_file).getroot()
    
    # Initialise the info dict 
    info_dict = {}
    info_dict['bboxes'] = []

    # Parse the XML Tree
    for elem in root:
        # Get the file name 
        if elem.tag == "filename":
            info_dict['filename'] = elem.text
            
        # Get the image size
        elif elem.tag == "size":
            image_size = []
            for subelem in elem:
                image_size.append(int(subelem.text))
            
            info_dict['image_size'] = tuple(image_size)
        
        # Get details of the bounding box 
        elif elem.tag == "object":
            bbox = {}
            for subelem in elem:
                if subelem.tag == "name":
                    bbox["class"] = subelem.text
                    
                elif subelem.tag == "bndbox":
                    for subsubelem in subelem:
                        bbox[subsubelem.tag] = int(subsubelem.text)            
            info_dict['bboxes'].append(bbox)
    
    return info_dict
```

```python
extract_info_from_xml(os.path.join(main_dir, '190614104153-07-shibuya-crossing-restricted_jpg.rf.50249fa874556e0d423f57faff106b21.xml'))
```

<h3>Convert to YOLO</h3>

The following steps were performed in sequence:

- Perform mapping

- Transform the bbox co-ordinates as per the format required by YOLO v5 and write to file

- Perform sample visualisation

```python
# Dictionary that maps class names to IDs
class_name_to_id_mapping = {"pedestrian": 0,
                           "car": 1}
```

```python
# Convert the info dict to the required yolo format and write it to disk
def convert_to_yolov5(info_dict):
    print_buffer = []
    
    # For each bounding box
    for b in info_dict["bboxes"]:
        try:
            class_id = class_name_to_id_mapping[b["class"]]
        except KeyError:
            print("Invalid Class. Must be one from ", class_name_to_id_mapping.keys())
        
        # Transform the bbox co-ordinates as per the format required by YOLO v5
        b_center_x = (b["xmin"] + b["xmax"]) / 2 
        b_center_y = (b["ymin"] + b["ymax"]) / 2
        b_width    = (b["xmax"] - b["xmin"])
        b_height   = (b["ymax"] - b["ymin"])
        
        # Normalise the co-ordinates by the dimensions of the image
        image_w, image_h, image_c = info_dict["image_size"]  
        b_center_x /= image_w 
        b_center_y /= image_h 
        b_width    /= image_w 
        b_height   /= image_h 
        
        #Write the bbox details to the file 
        print_buffer.append("{} {:.3f} {:.3f} {:.3f} {:.3f}".format(class_id, b_center_x, b_center_y, b_width, b_height))
        
    # Name of the file which we have to save 
    save_file_name = os.path.join(main_dir, info_dict["filename"].replace(".jpg", ".txt"))
    
    # Save the annotation to disk
    print("\n".join(print_buffer), file = open(save_file_name, "w"))
```

```python
# Get the annotations
annotations = [os.path.join(main_dir, x) for x in os.listdir(main_dir) if x[-3:] == "xml"]
annotations.sort()

# Convert and save the annotations
for ann in tqdm(annotations):
    info_dict = extract_info_from_xml(ann)
    convert_to_yolov5(info_dict)
annotations = [os.path.join(main_dir, x) for x in os.listdir(main_dir) if x[-3:] == "txt"]
```

```python
random.seed(0)

class_id_to_name_mapping = dict(zip(class_name_to_id_mapping.values(), class_name_to_id_mapping.keys()))

def plot_bounding_box(image, annotation_list):
    annotations = np.array(annotation_list)
    w, h = image.size
    
    plotted_image = ImageDraw.Draw(image)

    transformed_annotations = np.copy(annotations)
    transformed_annotations[:,[1,3]] = annotations[:,[1,3]] * w
    transformed_annotations[:,[2,4]] = annotations[:,[2,4]] * h 
    
    transformed_annotations[:,1] = transformed_annotations[:,1] - (transformed_annotations[:,3] / 2)
    transformed_annotations[:,2] = transformed_annotations[:,2] - (transformed_annotations[:,4] / 2)
    transformed_annotations[:,3] = transformed_annotations[:,1] + transformed_annotations[:,3]
    transformed_annotations[:,4] = transformed_annotations[:,2] + transformed_annotations[:,4]
    
    for ann in transformed_annotations:
        obj_cls, x0, y0, x1, y1 = ann
        plotted_image.rectangle(((x0,y0), (x1,y1)))
        
        plotted_image.text((x0, y0 - 10), class_id_to_name_mapping[(int(obj_cls))])
    
    plt.imshow(np.array(image))
    plt.show()

# Get any random annotation file 
annotation_file = random.choice(annotations)
with open(annotation_file, "r") as file:
    annotation_list = file.read().split("\n")[:-1]
    annotation_list = [x.split(" ") for x in annotation_list]
    annotation_list = [[float(y) for y in x ] for x in annotation_list]

#Get the corresponding image file
image_file = annotation_file.replace("annotations", "images").replace("txt", "jpg")
assert os.path.exists(image_file)

#Load the image
image = Image.open(image_file)

#Plot the Bounding Box
plot_bounding_box(image, annotation_list)
```

<h3>Environment setup</h3>

The following steps were performed in sequence:

- Create images and labels folders, with train, validation, test sub-folders

- Create iteration loop to move images and labels

- Perform random shuffling

- Print os.listdir to confirm file list before shift

- Perform shift

- Perform clone of Yolov5 onto local environment and its requirements

```python
# Create two subfolders if they don't exist

images = os.path.join(main_dir, "images")
labels = os.path.join(main_dir, "labels")

os.makedirs(images, exist_ok = True)
os.makedirs(labels, exist_ok = True)
```

```python
# Iterate over files in the source directory
for file_name in os.listdir(main_dir):
    # Check if the file is a JPEG image
    if file_name.endswith(".jpg"):
        # Move the image file to the image subfolder
        shutil.move(os.path.join(main_dir, file_name), os.path.join(images, file_name))
    # Check if the file is a TXT annotation file
    elif file_name.endswith(".txt"):
        # Move the annotation file to the label subfolder
        shutil.move(os.path.join(main_dir, file_name), os.path.join(labels, file_name))
```

```python

# Create a folder structure for YOLOv5 training
if not os.path.exists('data'):
    for folder in ['images', 'labels']:
        for split in ['train', 'val', 'test']:
            os.makedirs(f'data/{folder}/{split}')
```

```python
def get_filenames(folder):
    filenames = set()
    
    for path in glob.glob(os.path.join(folder, '*.jpg')):
        # Extract the filename
        filename = os.path.split(path)[-1]        
        filenames.add(filename)

    return filenames
```

```python
image_proc = get_filenames('main/images')

images_process = np.array(list(image_proc))

# Use the same random seed for reproducability
np.random.seed(42)
np.random.shuffle(images_process)
```

```python
image_dir = 'main/images'
label_dir = 'main/labels'

image_name = os.listdir(image_dir)
label_name = os.listdir(label_dir)

print(image_name)
print(label_name)
```

Perform shift

```python
def split_dataset(image_names, train_size, val_size):
    image_dir = 'main/images'
    label_dir = 'main/labels'

    image_names = os.listdir(image_dir)
    
    for i, image_name in enumerate(image_names):
        # Label filename
        label_name = image_name.replace('.jpg', '.txt')
        
        # Split into train, val, or test
        if i < train_size:
            split = 'train'
        elif i < train_size + val_size:
            split = 'val'
        else:
            split = 'test'
        
        # Source paths
        source_image_path = os.path.join('main', 'images', image_name)
        source_label_path = os.path.join('main', 'labels', label_name)

        # Destination paths
        target_image_folder = os.path.join('data', 'images', split)
        target_label_folder = os.path.join('data', 'labels', split)

        # Print for debugging
        #print("Copying:", source_image_path, "to", target_image_folder)
        #print("Copying:", source_label_path, "to", target_label_folder)

        # Copy files
        shutil.copy(source_image_path, target_image_folder)
        shutil.copy(source_label_path, target_label_folder)

# Cat data
split_dataset(images, train_size=32, val_size=8)

```

<h2>Training</h2>

- pedestrian_and_car.yaml is the file containing directory pathways to train-val-test and the number of classes

- yolov5s.pt is the pre-trained weight; small weight chosen for faster training time and less use of memory resources

- epochs is the number of training epochs; test images with cars and/or pedestrians are variously fitted after 10, 30 and 50 epochs

- batch refers to training batch size; considering that there are only 40 images used together for training and validation, batch size of 4 is appropriate.

- Freeze refers to the layer number taken out from Yolov5 to be used for transfer learning. Yolov5 has a total of 24 layers, freeze 10 means that the first 10 layers of CNN are frozen. The first 10 layers constitutes the "backbone" of Yolov5.
 
- Path to the weight files can be found under: yolov5/runs/train/exp"x"/weights/best.pt, where x is the a number representing the number of run iteration
